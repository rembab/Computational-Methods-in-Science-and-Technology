# Matrix operations

```python
def multiply_row(row, mul, X):
    for col in range(len(X[row])):
        X[col][row] *= mul

def add_row(row1, row2, mul, X):
    for col in range(len(X[row1])):
        X[col][row1] += X[col][row2] * mul

def swap_rows(row1, row2, X):
    for col in range(len(X[row1])):
        X[col][row1], X[col][row2] = X[col][row2], X[col][row1]

def swap_cols(col1, col2, X):
    for row in range(len(X[col1])):
        X[col1][row], X[col2][row] = X[col2][row], X[col1][row]

def print_matrix(X):
    N = len(X)
    for row in range(N):
        for col in range(N):
            print(int(X[col][row]), end=" ")
        print()
```

# Random matrices and vectors for testing purposes

```python

def random_matrix(N):
    return [[random.randrange(0, 10) for row in range(N)] for col in range(N)]


def random_vector(N):
    return [random.randrange(0, 10) for row in range(N)]
```

# LU factorization
Heavily similiar to Gauss elimination. All of the factorization is done in place. Additional function for creating two distinct arrays.

```python
def LU_factorization(Xin):
    X = copy.deepcopy(Xin)
    N = len(X)

    for col in range(N - 1):
        for row in range(col + 1, N, 1):
            if X[col][col] == 0:
                return None
            mul = X[row][col] / X[col][col]
            X[row][col] = mul
            for c in range(col + 1, N):
                X[row][c] -= mul * X[col][c]
                
    return X

def LU(Xin):
    combined = LU_factorization(Xin)
    if not combined: return None, None
    N = len(Xin)
    
    lc = lambda r, c : 1.0 if r == c else (combined[r][c] if r > c else 0)
    uc = lambda r, c : combined[r][c] if r <= c else 0
    L = [[lc(r, c) for c in range(N)] for r in range(N)]
    U = [[uc(r, c) for c in range(N)] for r in range(N)]
    return L, U
```

# Testing

As expected there is numerical instability due to lack of pivoting, even including occasional potential zero division errors

```python
X = [[1,4,7],[2,5,8],[3,6,9]]
L, U = LU(X)
Ns = list(range(50,501,50))
for N in Ns:
    A = random_matrix(N)    
    L,U = LU(A)
    print("Testing for N =",N)
    if not L: 
        print("Skipped due to zero pivot")
        continue
    print("||A - LU|| = ",np.linalg.norm(A - np.dot(L,U)))
```
